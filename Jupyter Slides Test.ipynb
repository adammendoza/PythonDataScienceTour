```python
import pandas
import numpy
import os
%matplotlib inline
```

```python
df_train = pandas.read_csv(os.path.join('/Users/cls/workspace/foodtracks/forecasting_prepared_anonym_benchmark.zip'), sep=';',
                       index_col=0, parse_dates=['date'], infer_datetime_format=True)
```

```python
# transform sales table to disentangle sales time series
sales = df_train[["date", "store_id", "article_id", "amount_sold"]]
sales.index = pandas.DatetimeIndex(sales["date"])
sales = sales.pivot_table(index=["store_id", "article_id", "date"])
```
